```python
from statsmodels.formula.api import ols
```

```python
.columns = [col.replace(' ', '_') for col in kc.columns]
for col in x_cols:
    kc[col] = (kc[col] - kc[col].mean())/kc[col].std()
kc.head()
```

```python
predictors = '+'.join(x_cols)
formula = outcome + '~' + predictors
model = ols(formula=formula, data=kc).fit()
model.summary()
```
